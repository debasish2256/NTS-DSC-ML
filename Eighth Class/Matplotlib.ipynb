# Matplot

```python
import matplotlib.pyplot as plt
from matplotlib import style

#style.use("fivethirtyeight")#for style
style.use("ggplot")#style



x1=[5,8,10]
y1=[12,16,6]


x2=[6,9,11]
y2=[5,1,7]

#plt.plot(x,y)
plt.plot(x1,y1,linewidth=5,color="g",label="1st one")
plt.plot(x2,y2,linewidth=5,color="c",label="2nd one")


plt.title("My Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.grid(color="k")#used for coloring grid

plt.legend()#used for naming line
plt.show()
```

# Bar Graph

```python
import matplotlib.pyplot as plt


plt.bar([1,3,5,7,9],[5,2,7,8,2],color="c",label="ages")
plt.bar([2,4,6,8,10],[8,6,2,5,6],color="g",label="number")


plt.xlabel("My x")
plt.ylabel("My y")

plt.title("my 1st bar")
plt.axis([0,12,0,10])#for fixig length of graph
plt.legend()

plt.show()
```

# Stack/Area plot

```python


days=[1,2,3,4,5]

sleeping=[7,8,6,11,7]

eating=[2,3,4,3,2]

working=[7,8,7,2,2]


playing=[8,5,7,8,13]

#used for legend marking
plt.plot([],[],color="m",label="sleeping",linewidth=5)
plt.plot([],[],color="c",label="eating",linewidth=5)
plt.plot([],[],color="r",label="working",linewidth=5)
plt.plot([],[],color="g",label="playing",linewidth=5)


plt.stackplot(days,sleeping,eating,working,playing,colors=["m","c","r","g"])
plt.title("Area Plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

# Pie Chart

```python
slices=[7,5,12,13]
activities=["Sleeping","Eating","Working","Playing"]
cols=["c","m","r","b"]
            
plt.pie(slices,labels=activities,colors=cols,shadow=True,autopct="%1.2f%%",startangle=90,explode=(0,0.3,0,0)) #autopct used for marking percentage
#explode used for making different in pie chart
plt.title("Pie Chart")
plt.show()
            
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

#arnge(start,end,interval)
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)




#subplot(nrows,ncols,plot_number)

#where nrows and ncols are used to notionally split the figure into nrows*ncols sub-axes,and plot_number is use to identify the particular
#subplot that thios function is to create wiyhin the notional grid.plot_number start at 1,increaments across rows first and has a maximum of nrows * ncols


plt.subplot(211)
plt.plot(t1,f(t1),"bo",t2,f(t2)) #bo used for dotted purpose

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2))
plt.show()
```
