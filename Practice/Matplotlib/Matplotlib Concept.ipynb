```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
#Functional
plt.plot(x,y)
plt.title("Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"black")
plt.subplot(1,2,2)
plt.plot(y,x,"b")
```

```python
##00
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_title("Set Title")
axes.set_xlabel("X-axis")
axes.set_ylabel("Y-axis")
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("Larger Plot")

axes2.plot(y,x)
axes2.set_title("Smaller Plot")
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)


#axes.plot(x,y)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title("First plot")

axes[1].plot(y,x)
axes[1].set_title("second plot")
```

# Fig.  size  and DPI

```python
fig=plt.figure(figsize=(8,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="X-squared")
ax.plot(x,x**3,label="X-Cubed")
ax.legend(loc=0)
```

# Plot Appearence

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color="Purple",lw=3,ls="steps",marker="o",markersize=10,
        markerfacecolor="yellow",markeredgewidth=3,
        markeredgecolor="green")
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color="Purple",lw=3,ls="-",marker="o",markersize=10,
        markerfacecolor="yellow",markeredgewidth=3,
        markeredgecolor="green")
```
