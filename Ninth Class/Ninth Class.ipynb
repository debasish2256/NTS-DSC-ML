```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  #advance version of matplotlib

%matplotlib inline 
#output the graph in the same shell
```

```python
#Joint Plot

iris=sns.load_dataset("iris")
print(iris)

iris.head()
```

```python
iris["species"].value_counts()
```

```python
#Joint Plot

iris=sns.load_dataset('iris')
sns.jointplot(x='sepal_length',y='sepal_width',data=iris)
```

```python
#Joint Plot

iris=sns.load_dataset('iris')
sns.jointplot(x='petal_length',y='petal_width',data=iris)
```

```python
#Faced Grd Plot
#\ used for join


sns.FacetGrid(iris,hue="species",size=5)\
.map(plt.scatter,"sepal_length","sepal_width")\
.add_legend()
```

```python
#bar graph

sns.boxplot(x="species",y="petal_length",data=iris)
```

```python
#violin plot

sns.violinplot(x="species",y="petal_length",data=iris,size=5)
```

```python
#Pair Plot

sns.pairplot(data=iris,kind="scatter")
```

```python
#Heat Map

df=pd.DataFrame(np.random.rand(5,5),columns=["a","b","c","d","e"])
print(df)
```

```python
corr=df.corr()
print(corr)
```

```python
sns.heatmap(corr)
```

```python
flights=sns.load_dataset("flights")
print(flights)
flights.head()
```

```python
flights=flights.pivot("month","year","passengers")
sns.heatmap(flights)
```

```python
sns.heatmap(flights,annot=True,fmt="d",linewidth=.5)
#annnot gives numerical value to each correlation
```

```python
sns.heatmap(flights,annot=True,fmt="d",linewidth=.5,cmap="YlGnBu")
```

```python
#Dist Plot

x=np.random.normal(size=100)
print(x)
sns.distplot(x)
```

```python
    sns.distplot(x,kde=False,rug=True)
```

```python
sns.distplot(x,bins=30,kde=False,rug=True)
```

```python
#Kernal density estimation
#it is useful tool for plotting the shape of distribution
print(sns.distplot(x,hist=False,rug=True))
```
